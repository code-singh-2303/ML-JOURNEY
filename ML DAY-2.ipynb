```python
import numpy as np
import matplotlib.pyplot as plt
def f(x):
    return x**2

def derivative(f,x,h=0.000000000001):
    return ((f(x+h)-f(x))/h)

print(derivative(f,2))
print(derivative(f,4))


x = np.linspace(-5,5,20)
y = f(x)
dy = derivative(f,x)

plt.Figure(figsize=(8,4))
plt.plot(x,y,label='f(x)=x^2')
plt.plot(x,dy,label="f'(x)=2x")
plt.legend
plt.show
plt.grid(True)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2 + 4*x + 4
def df(x):
    return 2*x + 4

#gradient descent 
x=10 
lr = 0.01
steps = []
for i in range(50):
    steps.append(x)
    grad = df(x)
    x = x - lr*grad
print(f"Minimum found at x= {x:.4f}")
x_vals = np.linspace(-15, 15, 100)
plt.plot(x_vals, f(x_vals), label='f(x)')
plt.scatter(steps, [f(s) for s in steps], c='red', s=20, label='descent steps')
plt.legend()
plt.show()
```
