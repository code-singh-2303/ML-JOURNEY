```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import os

os.chdir(r'C:\Users\aruna\Desktop\Work\MLDAY1')

df = pd.read_csv('Titanic-Dataset.csv')
print(df.shape)
print(df.isnull().sum())
```

```python
df['Age']=df['Age'].fillna(df['Age'].median())
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
print(df.isnull().sum())
```

```python
df.drop(columns=['PassengerId','Name','Ticket'],inplace=True)
print(df.columns)
```

```python
df['Sex'] = df['Sex'].map({'male':0,'female':1})
df['Embarked'] = df['Embarked'].map({'S':0,'C':1,'Q':2})
print(df.head())
print(df.dtypes)
```

```python
df['FamilySize']=df['SibSp']+df['Parch']+1
df['IsAlone']=(df['FamilySize']==1).astype(int)
print(df[['SibSp', 'Parch', 'FamilySize', 'IsAlone']].head(10))
print(df.groupby('IsAlone')['Survived'].mean())
```

```python
# save cleaned dataset for use on Day 8 when you train your first model
df.to_csv('Titanic-Cleaned.csv', index=False)
print("saved successfully")
print(df.shape)   # should be (891, 9)
print(df.head())  # final look at clean data
```
