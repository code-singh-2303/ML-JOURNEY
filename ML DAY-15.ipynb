```python
import numpy as np

#activation function

def sigmoid(x):
    return 1/ (1+np.exp(-x))

def relu(x):
    return np.maximum(0,x)

#single neuron

def neuron(inputs,weights,bias,activation='relu'):
    z=np.dot(inputs,weights) + bias
    if activation=='relu':
        return relu(z)
    elif activation=='sigmoid':
        return sigmoid(z) 

#test

inputs=np.array([0.5,0.3,0.8])
weights=np.array([0.4,0.2,0.6])
bias=0.1

output_relu=neuron(inputs,weights,bias,activation='relu')
output_sigmoid = neuron(inputs,weights,bias,activation='sigmoid')

print(f"Raw z: {np.dot(inputs,weights)+bias:.4f}")
print(f"RELU Output: {output_relu:.4f}")
print(f"Sigmoid Output: {output_sigmoid: .4f}")




```

```python
# test with negative input
inputs_neg  = np.array([-0.5, -0.3, -0.8])
output_relu_neg    = neuron(inputs_neg, weights, bias, activation='relu')
output_sigmoid_neg = neuron(inputs_neg, weights, bias, activation='sigmoid')

print(f"Raw z:          {np.dot(inputs_neg, weights) + bias:.4f}")
print(f"ReLU output:    {output_relu_neg:.4f}")
print(f"Sigmoid output: {output_sigmoid_neg:.4f}")
```

```python
class NeuralNetwork:
    def __init__(self):
        # layer 1: 3 inputs → 4 neurons
        self.W1 = np.random.randn(3, 4) * 0.01
        self.b1 = np.zeros(4)
        
        # layer 2: 4 inputs → 2 outputs
        self.W2 = np.random.randn(4, 2) * 0.01
        self.b2 = np.zeros(2)
    
    def forward(self, X):
        # layer 1
        self.z1 = np.dot(X, self.W1) + self.b1
        self.a1 = relu(self.z1)
        
        # layer 2
        self.z2 = np.dot(self.a1, self.W2) + self.b2
        self.a2 = sigmoid(self.z2)
        
        return self.a2

# test forward pass
nn = NeuralNetwork()
X = np.array([[0.5, 0.3, 0.8]])  # one sample, 3 features
output = nn.forward(X)
print(f"Output shape: {output.shape}")
print(f"Output: {output}")
print(f"Sum: {output.sum():.4f}")  # won't sum to 1 yet — that needs softmax
```

```python
W1 = np.array([[0.1, 0.2, -0.1, 0.3],
               [0.4, -0.2, 0.1, 0.2],
               [-0.1, 0.3, 0.2, -0.3]])

b1 = np.array([0.1, 0.0, -0.1, 0.2])

W2 = np.array([[0.2, -0.1],
               [0.3, 0.2],
               [-0.2, 0.4],
               [0.1, -0.3]])

b2 = np.array([0.1, -0.1])

X = np.array([[0.5, 0.3, 0.8]])
nn.W1 = W1
nn.b1 = b1
nn.W2 = W2
nn.b2 = b2

output = nn.forward(X)
print(f"z1: {nn.z1}")
print(f"a1: {nn.a1}")
print(f"z2: {nn.z2}")
print(f"a2: {nn.a2}")
```
